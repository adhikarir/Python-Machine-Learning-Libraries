# Seaborn
Seaborn is a data visualization library built on top of matplotlib and closely integrated with pandas data structures in Python. Visualization is the central part of Seaborn which helps in exploration and understanding of data.

To initialize the Seaborn library, the command used is:

```python
import seaborn as sns
```

Using Seaborn we can plot wide varieties of plots like:
    Distribution Plots
    Pie Chart & Bar Chart
    Scatter Plots
    Pair Plots
    Heat maps

```python
#Plotting a Displot
import matplotlib.pyplot as plt
sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
#Plotting a Distplot Without the Histogram
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```

### 1.Distribution Plots
We used the googleplaystore data for the example, datasource link: https://www.kaggle.com/zeshanali/googleplaystore?select=googleplaystore.csv

```python
import pandas as pd
import numpy as np

pstore = pd.read_csv("data\googleplaystore.csv")
# The dataset looks like this,
pstore.head(10)
```

Let’s see how distribution plot looks like if we plot for ‘Rating’ column from the above dataset,

```python
#Create a distribution plot for rating
sns.distplot(pstore.Rating)
plt.show()
```

we can change the number of bins and make the graph more understandable. We just have to add the number of bins in the code,

```python
#Change the number of bins
sns.distplot(pstore.Rating, bins=20, kde = False)
plt.show()
```

We can also provide the title and color of the bins similar to matplotlib to the distribution plots. Let’s see the code for that,

```python
#Create a distribution plot for rating
sns.distplot(pstore.Rating, bins=20, color="g")
plt.title("Distribution of app ratings", fontsize=20, color = 'red')
plt.show()
```

#### Styling the Seaborn graphs

We just have to write one line of code to incorporate these styles into our graph.

```python
#Adding dark background to the graph
plt.style.use("dark_background")

#Create a distribution plot for rating
sns.distplot(pstore.Rating, bins=20, color="g")
plt.title("Distribution of app ratings", fontsize=20, color = 'red')
plt.show()
```

### 2. Pie Chart & Bar Chart
Pie Chart is generally used to analyze the data on how a numeric variable changes across different categories.

```python
#Analyzing the Content Rating column
pstore['Content Rating'].value_counts()
```

As per the above output, since the count of “Adults only 18+” and “Unrated” are significantly less compared to the others, we’ll drop those categories from the Content Rating and update the dataset.

```python
#Remove the rows with values which are less represented 
pstore = pstore[~pstore['Content Rating'].isin(["Adults only 18+","Unrated"])]

#Resetting the index
pstore.reset_index(inplace=True, drop=True)

#Analyzing the Content Rating column again
pstore['Content Rating'].value_counts()
```

#### Pie Chart for the categories present in the Content Rating column.

```python
#Plotting a pie chart
plt.figure(figsize=[9,7])
pstore['Content Rating'].value_counts().plot.pie()
plt.show()
```

#### Bar chart.

```python
#Plotting a bar chart
plt.figure(figsize=[9,7])
pstore['Content Rating'].value_counts().plot.barh()
plt.show()
```

### 3.Scatter Plots
Scatter Plot is used when we want to plot the relationship between any two numeric columns from a dataset. These plots are the most powerful visualization tools that are being used in the field of machine learning.

#### Scatter Plot using matplotlib

```python
#Plotting the scatter plot
plt.scatter(pstore.Size, pstore.Rating)
plt.show()
```

#### Scatter Plot using Seaborn
We will be using sns.joinplot() in the code for scatter plot along with the histogram.
sns.scatterplot() in the code for only scatter plots.

```python
# Plotting the same thing now using a jointplot
sns.jointplot(pstore.Size, pstore.Rating)
plt.show()
```

#### Regression Plot
Regression plots create a regression line between 2 numerical parameters in the jointplot(scatterplot) and help to visualize their linear relationships.

```python
# Plotting the same thing now using a jointplot
sns.jointplot(pstore.Size, pstore.Rating, kind = "reg")
plt.show()
```

### 4. Pair Plots
Pair Plots are used when we want to see the relationship pattern among more than 3 different numeric variables. For example, 
Let’s create a pair plot for Reviews, Size, Price, and Rating columns from of dataset.
We will be using sns.pairplot() in the code to plot multiple scatter plots at a time.

```python
# Plotting the same thing now using a jointplot
sns.pairplot(pstore[['Reviews', 'Size', 'Price','Rating']])
plt.show()
```
